```python
class Solution:
    def snakesAndLadders(self, moves: List[List[int]]) -> int:
        def snake_ladder(board):
            n = len(board)
            queue = deque()
            visited = [False for _ in range(n)]
            queue.append([1, 0])
            visited[1] = True
            moves = float('inf')
            while queue:
                i, move_no = queue.popleft()
                b = i
                for _ in range(7):
                    # print(b, queue)
                    if i == n:
                        break
                    if board[i] > -1:
                        a = board[i]
                    else:
                        a = i
                    if not visited[a]:
                        queue.append([a, move_no + 1])
                        visited[a] = True
                    if b == n - 1:
                        print(i, n - 1, move_no, moves)
                        moves = min(moves, move_no)
                        break
                    i += 1
            return moves
        board = [0]
        m = len(moves)
        n = len(moves[0])
        odd = True
        ladder = False
        for i in range(m - 1, -1, -1):
            if odd:
                for j in range(n):
                    board.append(max(-1, moves[i][j]))
                odd = False
            else:
                for j in range(n - 1, -1, -1):
                    board.append(max(-1, moves[i][j]))
                odd = True
        # print(board, len(board))
        res = snake_ladder(board)
        if res == float('inf'):
            return -1
        return res
```
