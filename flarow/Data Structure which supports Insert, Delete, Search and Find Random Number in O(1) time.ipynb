```python
import random
class Listionary:
    def __init__(self):
        self.list = []
        self.dict = {}

def insert(listionary, num, n):
    listionary.list.append(num)
    listionary.dict[num] = n
    n += 1
    return listionary, n

def display(listionary):
    print(listionary.list)

def delete(listionary, num, n):
    if num not in listionary.dict:
        return listionary, n, False
    idx = listionary.dict[num]
    del listionary.dict[num]
    n -= 1
    listionary.list[idx], listionary.list[-1] = listionary.list[-1], listionary.list[idx]
    listionary.list.pop()
    print(idx, n)
    if idx != n:
        val = listionary.list[idx]
        listionary.dict[val] = idx
    return listionary, n, True

def search(listionary, num):
    if num not in listionary.dict:
        return False
    return listionary.dict[num] + 1

def find_random_number(listionary):
    return random.choice(listionary.list)

if __name__=='__main__':
    listionary = Listionary()
    n = 0
    while True:
        print('Enter -1 to exit.')
        print('1. Insert a number.')
        print('2. Display the list.')
        print('3. Delete a number.')
        print('4. Search for a number.')
        print('5. Find a random number.')
        ch = int(input('Enter your choice: '))
        if ch == -1:
            break
        if ch == 1:
            num = int(input('Enter the number you want to insert: '))
            listionary, n = insert(listionary, num, n)
        if ch == 2:
            display(listionary)
        if ch == 3:
            num = int(input('Enter the number you want to delete: '))
            listionary, n, b = delete(listionary, num, n)
            if not b:
                print('Number not found in the list.')
        if ch == 4:
            num = int(input('Enter the number you want to search: '))
            idx = search(listionary, num)
            if idx:
                print(num, 'is present at index', idx - 1, 'in the list')
            else:
                print('Number not found in the list.')
        if ch == 5:
            num = find_random_number(listionary)
            print('Random number is', num)
```
