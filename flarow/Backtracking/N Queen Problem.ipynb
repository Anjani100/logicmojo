```python
def valid_path(board, row, col, n):
    for i in range(col):
        if board[row][i] == 1:
            return False
    i = row
    j = col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1
    i = row
    j = col
    while i < n and j >= 0:
        if board[i][j] == 1:
            return False
        i += 1
        j -= 1
    return True

def n_queen_problem(board, col, n):
    if col >= n:
        return True
    for row in range(n):
        if valid_path(board, row, col, n):
            board[row][col] = 1
            if n_queen_problem(board, col + 1, n):
                return True
            board[row][col] = 0
    return False

n = 4
board = [[0 for j in range(n)] for i in range(n)]
n_queen_problem(board, 0, n)
for b in board:
    print(b)
print()
```

### Printing All Possible Combinations

```python
def valid_path(board, row, col, n):
    for i in range(col):
        if board[row][i] == 1:
            return False
    i = row
    j = col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1
    i = row
    j = col
    while i < n and j >= 0:
        if board[i][j] == 1:
            return False
        i += 1
        j -= 1
    return True

def n_queen_problem(board, col, n):
    if col == n:
        for b in board:
            print(b)
        print()
        return True
    res = False
    for row in range(n):
        if valid_path(board, row, col, n):
            board[row][col] = 1
            res = n_queen_problem(board, col + 1, n) or res
            board[row][col] = 0
    return res

n = 8
board = [[0 for j in range(n)] for i in range(n)]
n_queen_problem(board, 0, n)
```
