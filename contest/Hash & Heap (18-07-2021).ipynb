### Another problem in hashing

```python
#!/bin/python3

import math
import os
import random
import re
import sys
from collections import defaultdict

#
# Complete the 'findCount' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER_ARRAY arr
#  2. INTEGER b
#

def findCount(arr, b):
    # Write your code here
    sum_dict = defaultdict(list)
    n = len(arr)
    res = set()
    for i in range(n - 1):
        for j in range(i + 1, n):
            s = arr[i] + arr[j]
            v = b - s
            for x, y in sum_dict.setdefault(v, []):
                if i == x or i == y or j == x or j == y:    continue
                res.add(tuple(sorted([arr[x], arr[y], arr[i], arr[j]])))
            sum_dict.setdefault(s, []).append([i, j])
    return len(res)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    arr = list(map(int, input().rstrip().split()))

    b = int(input().strip())

    result = findCount(arr, b)

    fptr.write(str(result) + '\n')

    fptr.close()
```

### Mapping the characters

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'findWindow' function below.
#
# The function is expected to return a STRING.
# The function accepts following parameters:
#  1. STRING a
#  2. STRING b
#

def findWindow(text, ptr):
    # Write your code here
    if len(text) < len(ptr):    return ''
    ascii_text = [0] * 256
    ascii_ptr = [0] * 256
    for p in ptr:   ascii_ptr[ord(p)] += 1
    start, count, min_window = 0, 0, float('inf')
    s = ''
    for i in range(len(text)):
        char = ord(text[i])
        ascii_text[char] += 1
        if ascii_ptr[char] and ascii_text[char] <= ascii_ptr[char]: count += 1
        if count == len(ptr):
            while ascii_text[ord(text[start])] > ascii_ptr[ord(text[start])] or not ascii_ptr[ord(text[start])]:
                ascii_text[ord(text[start])] -= 1
                start += 1
            if min_window > i - start:
                min_window = i - start
                s = text[start:i + 1]
    return s

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = input()

    b = input()

    result = findWindow(a, b)

    fptr.write(result + '\n')

    fptr.close()
```
