### Jesse and Cookies

```python
#!/bin/python3

import os
import sys
from heapq import heappush, heappop, heapify

#
# Complete the cookies function below.
#
def cookies(k, A):
    #
    # Write your code here.
    #
    heapify(A)
    count = 0
    while len(A) > 1 and A[0] < k:
        first = heappop(A)
        second = heappop(A)
        heappush(A, first + 2 * second)
        count += 1
    return count if A[0] >= k else -1

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    nk = input().split()

    n = int(nk[0])

    k = int(nk[1])

    A = list(map(int, input().rstrip().split()))

    result = cookies(k, A)

    fptr.write(str(result) + '\n')

    fptr.close()
```

### Find the Running Median

```python
#!/bin/python3

import os
import sys
from heapq import heappush, heappop, heapify

#
# Complete the runningMedian function below.
#
class FindMedian:
    def __init__(self):
        self.max_heap = []
        self.min_heap = []
        self.min_n = self.max_n = 0
    
    def add_num(self, num):
        if not self.min_n and not self.max_n:
            heappush(self.min_heap, -num)
            self.min_n += 1
            return
        if not self.max_n:
            if num < -self.min_heap[0]:
                heappush(self.max_heap, -heappop(self.min_heap))
                heappush(self.min_heap, -num)
            else:
                heappush(self.max_heap, num)
            self.max_n += 1
            return
        if num < -self.min_heap[0]:
            heappush(self.min_heap, -num)
            self.min_n += 1
        else:
            heappush(self.max_heap, num)
            self.max_n += 1
        if self.max_n - self.min_n > 1:
            heappush(self.min_heap, -heappop(self.max_heap))
            self.max_n -= 1
            self.min_n += 1
        if self.min_n - self.max_n > 1:
            heappush(self.max_heap, -heappop(self.min_heap))
            self.min_n -= 1
            self.max_n += 1
        
    def get_median(self):
        if self.min_n == self.max_n:
            return (self.max_heap[0] - self.min_heap[0]) / 2
        if self.min_n > self.max_n:
            return -self.min_heap[0]
        return self.max_heap[0]

def runningMedian(a):
    #
    # Write your code here.
    #
    median = FindMedian()
    res = []
    for elements in a:
        median.add_num(elements)
        res.append(median.get_median())
    return res


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a_count = int(input())

    a = []

    for _ in range(a_count):
        a_item = int(input())
        a.append(a_item)

    result = runningMedian(a)

    fptr.write('\n'.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```
