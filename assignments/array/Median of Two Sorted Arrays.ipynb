```python
# Time: O(n)
# Space: O(1)

def find_median(nums1, nums2):
    def helper(i, j, first):
        if i < m and j < n:
            second = min(nums1[i], nums2[j])
        elif i < m:
            second = nums1[i]
        else:
            second = nums2[j]
        if (m + n) % 2 == 0:
            return (first + second) / 2
        return second
    
    m, n = len(nums1), len(nums2)
    if not m:
        if n % 2:
            return nums2[n // 2]
        return (nums2[n // 2] + nums2[(n // 2) - 1]) / 2
    if not n:
        if m % 2:
            return nums1[m // 2]
        return (nums1[m // 2] + nums1[(m // 2) - 1]) / 2
    
    i = j = count = 0
    while i < m and j < n:
        if nums1[i] < nums2[j]:
            first = nums1[i]
            i += 1
        else:
            first = nums2[j]
            j += 1
        count += 1
        if count == (m + n) // 2:
            return helper(i, j, first)
    while i < m:
        i += 1
        count += 1
        if count == (m + n) // 2:
            return helper(i, j, nums1[i - 1])
    while j < n:
        j += 1
        count += 1
        if count == (m + n) // 2:
            return helper(i, j, nums2[j - 1])

if __name__=='__main__':
    tc = [[[1, 6, 8, 10, 12], [2, 3, 4, 9, 11]],
          [[1,2,3,6],[4,6,8,10]],
          [[2,7],[11,16]],
          [[0,0],[0,0]],
          [[1,2],[]],
          [[1,3],[2]],
          [[1,2],[3,4]],
          [[],[1]],
          [[2],[]],
          [[1],[2,3,4]]]
    for t1, t2 in tc:
        print(find_median(t1, t2))
```

```python
# Time: O(log(m + n))
# Space: O(1)

def findMedianSortedArrays(nums1, nums2):
    def find_median(nums1, nums2):
        m, n = len(nums1), len(nums2)
        low, high = 0, m
        while True:
            partition1 = max((low + high) // 2, 0)
            partition2 = (m + n + 1) // 2 - partition1
            x_left = nums1[partition1 - 1] if partition1 > 0 else float('-inf')
            x_right = nums1[partition1] if partition1 < m else float('inf')
            y_left = nums2[partition2 - 1] if partition2 > 0 else float('-inf')
            y_right = nums2[partition2] if partition2 < n else float('inf')
            if x_left <= y_right and y_left <= x_right:
                if (m + n) % 2: return max(x_left, y_left)
                return (max(x_left, y_left) + min(x_right, y_right)) / 2
            if y_left > x_right:
                low = partition1 + 1
            else:
                high = partition1 - 1

    if len(nums1) == len(nums2):    return find_median(nums1, nums2)
    return find_median(min(nums1, nums2, key = len), max(nums1, nums2, key = len))

if __name__=='__main__':
    tc = [[[1, 6, 8, 10, 12], [2, 3, 4, 9, 11]],
          [[1,2,3,6],[4,6,8,10]],
          [[2,7],[11,16]],
          [[0,0],[0,0]],
          [[1,2],[]],
          [[1,3],[2]],
          [[1,2],[3,4]],
          [[],[1]],
          [[2],[]],
          [[1],[2,3,4]]]
    for t1, t2 in tc:
        print(findMedianSortedArrays(t1, t2))
```
